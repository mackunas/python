## Standard Python Modules

A standard Python installation includes a library of standard modules.

A **module** is a file with the `.py` extension that contains a collection of functions, classes, variables, constants, and other Python objects used to solve specific tasks.

The contents of a module can be accessed and used by any other Python program.

You can display a list of the standard modules by running the following command:

```python
help("modules")
```

````python
# uncomment to see a list of the standard modules
# help('modules')
````

A module is imported using the following command:

```python
import module_name
```

Multiple modules can be imported at the same time:

```python
import module_name_1, module_name_2, ..., module_name_n
```

To call a function from a module, write the module name followed by a dot (`.`) and the function name:

```python
module_name.function_name()
```

````python
import math

print(math.factorial(5))      # 120
x = math.pi
print(math.cos(x))            # -1.0
````

````python
import math, random

print(math.degrees(math.asin(1)))  # 90.0
print(random.randint(-5, 5))  # random integer in [-5, 5]
````

A module can also be imported using a different name (an **alias**):

```python

## Programming style (PEP 8)
Most programming languages can be written (or, more accurately, formatted) in different styles. Some styles are easier to read than others. Writing code that is easy for others to understand is a good practice, and using a consistent coding style helps make programs more readable.

Most Python projects follow the style guidelines defined in **PEP 8**, which promote clear and readable code. Some of the most important recommendations are:

- Use **4 spaces** for indentation, and do not use tabs.
- Use blank lines to separate functions, classes, and larger blocks of code.
- Whenever possible, write comments on a separate line.
- Use **docstrings** to document functions, classes, and modules.
- Use spaces around operators and after commas, but not inside parentheses:

```python
a = f(1, 2) + g(3, 4)
```

- Use consistent naming conventions:
  - **Classes:** `CamelCase`
  - **Functions and methods:** `lower_case_with_underscores`
  - Always use `self` as the first parameter of instance methods.

Operations can be performed on parts of sequences using **slice notation**.

String slices are defined using two indexes separated by a colon:

```python
string[i : j]
```

**Important:** The character with index `j` (`string[j]`) is not included in the slice. The slice ends with the character `string[j-1]`.

Slice indexes can be omitted:
- If the first index is missing, `0` is used.
- If the second index is missing, the length of the string is used.

An index that is larger than the string length is replaced with the string length.

````python
print(string[0:10])
print(string[0:11])
print(string[2:7])
print(string[:5])
print(string[6:])
print(string[-5:]) # 5 last symbols
````

A string is an immutable sequence of characters. This means that individual characters in a string cannot be changed.

However, new strings can be created using methods and functions.

## Data Type conversion
A data type can be converted using the functions `int()`, `float()`, `bool()`, `complex()`, and `str()`.

### `int()`

````python
print(int(10.999))  # float > int
try:
    print(int(10 + 5j)) # complex > int
except TypeError as e:
    print('TypeError:', e)
print(int(True), int(False)) # bool > int
print(int('20')) # str > int
for bad in ['10.5', '0b1111']:
    try:
        print(int(bad))
    except ValueError as e:
        print('ValueError for', bad + ':', e)
````

### `float()`

````python
print(float(100)) #int > float
print(float(0b1111))
print(float(0o1234))
print(float(0xFACE3))
try:
    print(float(10 + 5j)) # complex > float
except TypeError as e:
    print('TypeError:', e)
print(float(True), float(False)) # bool > float
print(float('10')) # str > float
print(float('12.345'))
````

## Mutable and Immutable Objects

A **mutable object** is an object that can change its value while keeping the same `id()`.

An **immutable object** is an object with a fixed value.

If we want to change the value of an immutable object, a new object must be created.

### The `id()` Function

Every Python object is stored in a specific memory location.

The `id()` function returns the memory location of a given object. The address is returned as a base-10 number.

This function can be used to check whether objects refer to the same memory location, meaning they refer to the same data in memory.

### The `is` Keyword

`is` is a keyword that compares two objects and returns `True` if both objects have the same memory address.

`is not` returns `False` if both objects have the same memory address.

### Immutable Objects

- Numbers (`int`, `float`, `decimal`, `boolean`, `complex`)
- Strings
- Tuples
- Frozen sets

````python
a = 10
print('a =', a, ' id(a)=', id(a))
a = 20
print('a =', a, ' id(a)=', id(a))

a = 1000
b = a
c = a
print('id(a)=', id(a))
print('id(b)=', id(b))
print('id(c)=', id(c))
print('a is b ->', a is b)
print('a is c ->', a is c)
print('b is c ->', b is c)

a = 10
b = a
print('a is b ->', a is b)
a = 1000
print('a is b ->', a is b)

x = []
y = []
print('x == y ->', x == y)
print('x is y ->', x is y)
print('x is not y ->', x is not y)
````

Mutable objects include:

- Lists
- Sets
- Dictionaries

A **list** is a collection of values of any data type, separated by commas and enclosed in square brackets.

````python
my_list = [10, 20, 30, 40]
print(id(my_list))
my_list[2] = 3000
print(id(my_list))
print(my_list)

list1 = [10, 20, 30, 40]
list2 = list1
print('list1 is list2 ->', list1 is list2)
list1[0] = 100.52
print('list1 is list2 ->', list1 is list2)
print('list1 =', list1)
print('list2 =', list2)

list1 = [10, 20, 30, 40]
list2 = [10, 20, 30, 40]
print('id(list1)=', id(list1))
print('id(list2)=', id(list2))
print('list1 is list2 ->', list1 is list2)

t = (1, 2, 3)
try:
    t[0] = 99
except TypeError as e:
    print('TypeError:', e)

s = {1, 2, 2, 3}
s.add(4)
print('set:', s)
d = {'a': 1, 'b': 2}
d['c'] = 3
print('dict:', d)
fs = frozenset([1, 2, 2, 3])
print('frozenset:', fs)
````

## Input and Output

### Data Input

Data is entered from the keyboard using the `input()` function.

All data read from the keyboard is treated as strings (`<class 'str'>`).

To avoid calculation errors, the input data type must be converted during input or before performing calculations.

````python
a = (input('First number: '))
b = (input('Second number: '))
print('a + b =', int(a)+int(b))

a = int(input('First number: '))
b = int(input('Second number: '))
print('a + b =', a+b)
````

### The `eval()` Function

````python
b = bool(input('Enter True or False: '))
print('b =', b)
````

When converting to the Boolean (`bool`) type, all strings except empty strings (`''`) are considered `True`. Therefore, in this case, the `eval()` function is used for type conversion:

````python
b = eval(input(' Enter True or False: '))
print('b =', b)
b = eval(input('Enter Expression: '))
print('b =', b)
````

How to read multiple values from the keyboard in one line:

````python
s = (input('Enter 3 numbers: '))# read str() type object
print(s)
L = s.split() # splits values, craetes list() type object
print(L)
L_int = [int(x) for x in L]#str > int
print(L_int)
a, b, c = L_int
print('a =', a)
print('b =', b)
print('c =', c)
````

All performed steps are combined into a single line:

````python
a, b, c = [int(x) for x in (input('Enter 3 numbers: ')).split()]
print('a =', a)
print('b =', b)
print('c =', c)

a, b, c = [eval(x) for x in (input('Enter some values: ')).split(',')]
print('a =', a, type(a))
print('b =', b, type(b))
print('c =', c, type(c))
````

## Formatted output
### The String Formatting Method `.format()`

````python
first_name = 'John'
last_name = 'Smith'
salary = 2500
print('Hello {} {}. Your salary is {}'.format(first_name, last_name, salary))
print('Hello {0} {1}. Your salary is {2}'.format(first_name, last_name, salary))
print('Hello {1} {0}. Your salary is {2}'.format(first_name, last_name, salary))
print('Hello {a} {b}. Your salary is {c}'.format(a=first_name, b=last_name, c=salary))
print('Hello {} {}. Your salary is {}'.format('John', 'Smith', 2500))
````

### Output Formats

- `:d` – integer (`int`)
- `:f` – floating-point number (`float`)
- `:e` – exponential notation
- `:s` – string (`str`)

````python
n = 458.541315
try:
    print('{:d}'.format(n))  # wrong: n is float
except ValueError as e:
    print('ValueError:', e)
print('{:d}'.format(int(n)))
print('{:f}'.format(n))
print('{0:15f}'.format(n))
print('{0:15e}'.format(n))
print('{0:15.3f}'.format(n))
````

### Alignment

- `<` – align to the left
- `>` – align to the right
- `^` – align to the center

````python
print('{0:15f}{1}'.format(n, '|'))
print('{0:<15f}{1}'.format(n, '|'))
print('{0:>15f}{1}'.format(n, '|'))
print('{0:^15.3f}{1}'.format(n, '|'))
print('{:s}'.format('hello'))
````
